```python

# importing pandas module  
import pandas as pd 
import numpy as np
   
# reading csv file from url  
data = pd.read_csv("https://cdncontribute.geeksforgeeks.org/wp-content/uploads/nba.csv") 
  
# dropping null value columns to avoid errors 
#data.dropna(inplace = True) 
  

  
```

# Append to existing data frame with split value columns 

```python
data [["First_Name", "Last_Name"]] = data["Name"].str.split(" ", n = 1, expand = True)
data.head(3)
```

# Replace NaN with None for specific fields

```python
data["Name"].fillna("Name Not Available", inplace =True)
```

```python
data.head()
```

```python
data.groupby(by=['Name'])['Salary'].sum().head(3)
```

# Get the Sum of Salary, along with Team, Age whose Name starts with John 

```python
data[data["Name"].str.startswith ("John")].groupby(['Name','Team', 'Age', 'Height'])["Salary"].sum()#['Salary']
```

# This would give Sum of all measures group on specific fields....

```python
data[data["Name"].str.startswith ("John")].groupby(['Name','Team', 'Age', 'Height']).sum()#['Salary']
```

# Replace NaN with None for all fields

```python
data.fillna("None", inplace =True)
```

# Get the aggregation on different fields with different Group fields

```python
data.groupby(["Team", "Name"]).agg({"Number": np.mean, "Age" : np.max, "Weight" : np.max, "Salary" : np.sum}).head(10)
```
